# Citi Bike Data 2017

```python
# Dependencies & Setup
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# View raw data files to aggregate
os.listdir('Data')
```

```python
# Tableau Public can work with a maximum of only 15,000,000 rows. The total rows for the full year of 
# 2017 is 20,523,930 so only the summer and winter months data will be used
# Read Citi Bike data into Pandas data frames
jan = pd.read_csv('Data/JC-201701-citibike-tripdata.csv/JC-201701-citibike-tripdata.csv')
feb = pd.read_csv('Data/JC-201702-citibike-tripdata.csv/JC-201702-citibike-tripdata.csv')

jun = pd.read_csv('Data/JC-201706-citibike-tripdata.csv/JC-201706-citibike-tripdata.csv')
jul = pd.read_csv('Data/JC-201707-citibike-tripdata.csv/JC-201707-citibike-tripdata.csv')
aug = pd.read_csv('Data/JC-201708 citibike-tripdata.csv/JC-201708 citibike-tripdata.csv')
dec = pd.read_csv('Data/JC-201712-citibike-tripdata.csv/JC-201712-citibike-tripdata.csv')
```

```python
jan.columns = jun.columns
```

```python
feb.columns = jun.columns
```

```python
all_months = pd.concat([jan, feb, jun, jul, aug, dec])
all_months
```

```python
all_months["starttime"] = pd.to_datetime(all_months["starttime"])
all_months["stoptime"] = pd.to_datetime(all_months["stoptime"])
```

```python
all_months.dtypes
```

```python
all_months.gender.value_counts()
```

```python
def to_gender_name(gender):
    if gender == 1:
        return "MALE"
    elif gender == 2:
        return "FEMALE"
    else:
        return "OTHER"

# all_months["gender"].map(lambda gender: "MALE" if gender == 1 else "FEMALE" if gender == 2 else "OTHER")
all_months["gender"] = all_months["gender"].map(to_gender_name)
all_months
```

```python
all_months.to_csv("citibike_final.csv", index=False)
```
